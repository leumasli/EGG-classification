```
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, Sequential
from keras.layers import *
%cd /content/drive/Shareddrives/EE147/
```

```
dir = "/content/drive/Shareddrives/EE147/project_data/"
X_test = np.load(dir+"X_test.npy")
y_test = np.load(dir+"y_test.npy")
person_train_valid = np.load(dir+"person_train_valid.npy")
X_train_valid = np.load(dir+"X_train_valid.npy")
y_train_valid = np.load(dir+"y_train_valid.npy")
person_test = np.load(dir+"person_test.npy")

# adjust labels 
y_train_valid -= 769
y_test -= 769

y_train_valid = keras.utils.to_categorical(y_train_valid, 4)
y_test = keras.utils.to_categorical(y_test, 4)

X_train_valid = np.swapaxes(X_train_valid, 1, 2)
X_test = np.swapaxes(X_test, 1, 2)
print(X_train_valid.shape, y_train_valid.shape)
print(y_train_valid[0])
```

```
model = Sequential([Input(shape=(1000, 22)),
                    LSTM(22, dropout=0.2, return_sequences=True),
                    GRU(22, dropout=0.2),
                    Dense(4)])
loss = keras.losses.CategoricalCrossentropy()
model.compile(loss=loss,
              optimizer = 'adam',
            #   metrics = ['accuracy'])
              metrics=keras.metrics.CategoricalAccuracy())

model.fit(X_train_valid, y_train_valid, epochs=10, batch_size=50)
```
